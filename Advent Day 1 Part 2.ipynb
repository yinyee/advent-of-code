```python
# open file for reading

f = open('1_input', 'r')
```

```python
# read file into an array of lines
# doing it this way to avoid appending a newline character to the end of each line

lines = f.read().splitlines()
```

```python
# get the first (and only) line

line = lines[0]
```

```python
# store the length of the line

modulo = len(line)
```

```python
# find the distance halfway around the array
# // operator means integer division

halfway = modulo // 2
```

```python
# now we cannot just append to the back of the list
# we use modulo arithmetic (i.e. clock arithmetic) instead 
# https://en.wikipedia.org/wiki/Modular_arithmetic

keep = []

for i in range(len(line)):
    check = (i + halfway) % modulo
    print("{}: {} {}: {}".format(i, line[i], check, line[check]))
    if line[i] == line[check]:
        keep.append(line[i])
```

```python
# convert all the strings into integers

keep = [int(i) for i in keep]
```

```python
# finally print out the sum of the values in the array

print(sum(keep))
```
